## Definition of Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and data manipulation library built on top of the Python programming language. It is a popular library for data analysis and data manipulation. It is widely used for data analysis in data science and machine learning


### What is a DataFrame?


The DataFrame is a two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns.


### What is a Series?

A Series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively referred to as the index.

## Fetching Columns and Rows


```python
# Fetch the data of a particular column
toss = data["toss_winner"]
toss.head(3)
```

```python
# Fetch Multiple Columns data
toss_decision = data[["team1", "team2", "toss_winner", "toss_decision"]]
toss_decision.head(3)
```

```python
toss_decision.iloc[2]
# The above code will give the data of 2nd row
```

```python
# Get data from 2 to 5 rows
toss_decision.iloc[2:5]
```

```python
# Get data of 1,5,11 rows -> also called as Fancy Indexing

toss_decision.iloc[[1, 5, 11]]
```
