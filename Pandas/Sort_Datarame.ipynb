### Sort the Dataframe


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("CSV2.csv")
df
```

```python
# Sort the Dataframe according to the Role Column
df.sort_values("Role")
```

```python
# Sort the Dataframe according to the Name Column
df.sort_values("Name")
```

```python
# Add some columns in Dataframe
maths = np.random.randint(40, 101, size=10)
os = np.random.randint(50, 101, size=10)
dsa = np.random.randint(30, 90, size=10)

# Adding The values in the Dataframe

df["Maths"] = maths
df["Operating System"] = os
df["DSA"] = dsa

df
```

```python
# sort Descending or Ascending by one column
df.sort_values("Maths", ascending=False)

# Sort Ascending
df.sort_values("Maths")
```

```python
# Sort Ascending by Multiple Columns
df.sort_values(["Operating System", "Name", "DSA"], ascending=False)
```

```python
# Sort descending by multiple columns and Update the dataframe
df.sort_values(["Operating System", "Name", "DSA"], ascending=False,inplace=True)
df
```
