#### Compare Two CSV Files for Differences

```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("CSV1.csv",encoding='ISO-8859-1')
data1
```

```python
data2 = pd.read_csv("CSV2.csv",encoding='ISO-8859-1',index_col=False)
data2
```

```python
# Get the difference of Two Files

df1 = pd.read_csv('CSV1.csv')
df2 = pd.read_csv('CSV2.csv')

# Compare the DataFrames
differences = df1.compare(df2)

# Print the differences
print("Differences between file1.csv and file2.csv:")
print(differences)
```
