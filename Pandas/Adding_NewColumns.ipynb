### Adding New Columns

#### 1.2 Add New column with an array


```python
# First Import Numpy
# Creating an array of size 10(as Number of rows of File )

Marks = [32, 34, 45, 65, 76, 87, 89, 90, 43, 54]
DataFr["OOPS"] = Marks
DataFr
```

```python
# Create a OS marks array with Random int
# np.random.randint(start(inclusive), last(exclusive),size)

os = np.random.randint(40, 101, size=10)
os
```

```python
DataFr["OS Marks"] = os
DataFr
```

```python
# Create Random Float Number between 1 - 100

coa = np.random.uniform(40,101,size=10)
```
