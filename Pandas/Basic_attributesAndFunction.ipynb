## Attributes and Functions and Methods


#### Attributes

A value associated with an object with an (referenced by name using dotted expressions) is called an attribute.

Ex - df.columns


#### Functions

A group of related statements that perform a specific task.

Python built-in functions

1. max()
2. min()
3. len()
4. type()


#### Methods

A function that is defined inside a class is called a method.

Ex - df.head()
