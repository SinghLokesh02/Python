### Python Numpy

#### The python numpy library is used for data science and in ML. It is used because of the following reasons. 

*  Numpy provide efficient Storage 
*  It also provide better ways of handling data for processing 
*  It is fast 
*  It is easy to learn 
*  Numpy uses relatively less memory to store data

```python
import numpy as np
```

```python
myarr = np.array([[23,43,54,65,76],[2,453,4,745,5]],np.int16)
```

```python
print(myarr)
myarr
```

```python
# Printing the elements of the array by index

print(myarr[1,1])

# Get the number of rows and columns in array
print("The rows and columns in myarray is : ",myarr.shape)
```

```python
myarr.dtype
```

```python
# Change the array element by index
print("Printing the data at index before changing : ",myarr[1,2])
myarr[1,2] = 567
print("Printing the data at index before changing : ",myarr[1,2])
```

## Numpy Axis

```python
arr = [[1,2,3],[4,5,6],[7,8,9]]
```

```python
arr
```

```python
x = np.array(arr)
```

```python
x
```

```python
# It will transpose the array
x.T
```

```python
# Flat the 2-D Array
x.flat
```

```python
for item in x.flat:
    print(item)
```

```python
# get The dimension of the array
print('The dimension of array x is : ',x.ndim)
```

```python
# Get all element count in the array
x.size
```

```python
# Total bytes consumed by the array
print("Total bytes consumed by the array x is : ",x.nbytes)
```

### Mathematical Functions by numpy 

```python
arr1 = np.array([[1,2,3],
                 [32,4,5],
                 [54,65,76]])
```

```python
arr1
```

```python
arr2 = np.array([[12,5,3],
                 [3,4,5],
                 [5,9,74]])
```

```python
arr2
```

```python
# Print the sum of two 2-D arrays
arr1+arr2
```

```python
# Print the Product of two 2-D arrays
arr1*arr2
```

```python
np.sqrt(arr1)
```

#### Print the sum of all the matrix Element

```python
arr2.sum()
```

#### Print the Max of two 2-D arrays

```python
arr2.max()
```

#### Print the Min of two 2-D arrays

```python
arr2.min()
```

```python
# return the count of non zero element from the array
np.count_nonzero(arr2)
```
